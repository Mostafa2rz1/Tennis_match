```python
import pandas as pd
import pyarrow.parquet as pq
import os
```

read file in each days

```python
folders_of_table = "Desktop/Tennis_project/202405/20240501/data/raw"
folders_path = [os.path.join(folders_of_table, file) for file in os.listdir(folders_of_table) ]
folders_path
```

this method is expierd

```python
dfs = {}
i = 0
for folder in folders_path:
    
    file_paths = [os.path.join(folder, file) for file in os.listdir(folder) ]
    df = [pd.read_parquet(file, engine= 'pyarrow') for file in file_paths ]
    result_df = pd.concat([df[i] for i in range(len(df))], ignore_index=True)
    dfs[f'df_{i}'] = result_df
    i += 1

for key, value in dfs.items():
    print(f'{key}:')
    print(value)
    print('\n')

```
