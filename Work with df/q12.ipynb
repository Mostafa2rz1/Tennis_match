```python
import numpy as np
import pandas as pd
```

```python
#open player dataframe
df_home_team_1 = pd.read_csv('/home/mostafa2rz/Desktop/Tennis_project/202405/df_home_team_1.csv').drop_duplicates(subset='match_id')
df_away_team_1 = pd.read_csv('/home/mostafa2rz/Desktop/Tennis_project/202405/df_away_team_1.csv').drop_duplicates(subset='match_id')
df_away_team_1
```

```python
#open  2 colums of player dataframe 

df_gender_away =df_away_team_1[['gender', 'match_id']] 
df_gender_home =df_home_team_1[['gender', 'match_id']] 
#concate home and away dataframe 
df_gender_2x = pd.concat([df_gender_away,df_gender_home] )
df_gender_2x
```

```python
df_gender= df_gender_2x.drop_duplicates(subset='match_id')
df_gender
```

number of set for each match_id

```python
df_pbp = pd.read_csv('/home/mostafa2rz/Desktop/Tennis_project/202405/df_pbp.csv')
df_number_of_game_for_each_set = df_pbp.groupby(['match_id','set_id'])['game_id'].count().reset_index()
df_number_of_game_for_each_set 
```

```python
df_number_of_game_and_gender = df_number_of_game_for_each_set.merge(df_gender , on='match_id', how='left')
df_number_of_game_and_gender
```

```python
df_meadin_number_of_game_for_each_set_numerical_and_gender = df_number_of_game_and_gender.groupby(['gender','match_id','set_id'])['game_id'].median().unstack()
#behtare inja ba nemodar neshon bdim
df_meadin_number_of_game_for_each_set_numerical_and_gender 
```

```python
df_meadin_number_of_game_for_each_set_and_gender =df_number_of_game_and_gender.groupby('gender')['game_id'].median().reset_index()
df_meadin_number_of_game_for_each_set_and_gender
```

```python
df_number_of_game =df_number_of_game_for_each_set.groupby('match_id')['game_id'].sum().reset_index()
df_number_of_game_and_gender_1 = df_number_of_game.merge(df_gender , on='match_id', how='left')
df_number_of_game_and_gender_1.groupby('gender')['game_id'].median().reset_index()
```
